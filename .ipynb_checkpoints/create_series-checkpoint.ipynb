# Pandas Series01

So, first in order to use the Series defined in the pandas library we need to import it to our notebook.
the first line of code imports pandas and numpy to this notebook.

```python
import pandas as pd
import numpy as np
```

## Series using NumPy-array:
   

```python
array = np.array([1,2,3,4,5,'string',[0,1,0,2]])
```

This is how we can define an NumPy array.
Basically, we can give any list type object to the argument of 'np.array()' method and it will generate a NumPy array.
NumPy arrays are very useful in storing the outputs generated by scientific computations. and pandas can help us manipulating the data stored inside these array objects.

```python
series1 = pd.Series(array)
```

Here 'series1' is an object of type pandas-series.
To create a pandas series from an numpy array we can use the array name as the argument to the 'pd.Series()' method.
This line of code will create a series binded with name 'series1' and will assign the data found inside the array to series1.
So, how about printing this series!??

```python
series1
```

The name of the series (here, 'series1') works as a keyword and prints that series.
As we can see from the output of the line-6, Pandas series index starts from 0.

## Using Lists

We can also create series from python list objects in a very similar manner how we created series using array objects from numpy

```python
list1 = [1,2,3,4,5,'python',np.nan,'data-science']
```

In order to create a series using python lists consider the list 'list1'.

```python
series2 = pd.Series(list1)
```

As, we did in the case of numpy array. Here, also we just argument the 'pd.Series()' method with the list name here 'list1'.

So, let's try to print this series

```python
series2
```

As you might have expected it just created a series representing the data stored inside the 'list1'

## Using Dictionary

```python
We can also create pandas series using python dictionaries. Consider the dict1 in the following line
```

```python
dict1 = {
    '@': 'Symbol',
    '2': 12,
    22 : 'twenty-two'
}
```

```python
series3 = pd.Series(dict1)
```

This will create a series using the argumented dictionary name.

Let's create another series using this exact same python dictionary but using a little different syntax.

```python
series4 = pd.Series({
  '@' : 'Symbol',
  '2' : 12,
   22 : "twenty-two" 
})
```

Let's try to print the contents of 'series3' and 'series4' one by one.

```python
series3
```

```python
series4
```

See, the outputs of line 18 and line 19 closely, they are indentical.
